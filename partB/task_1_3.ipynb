# Task 1.3 — What the Paper Claims to Improve (9 marks)

**Paper**: *Robust Point Set Registration Using Gaussian Mixture Models* — Bing Jian and Baba C. Vemuri, IEEE TPAMI, 2011.

---

## Main Baseline: Iterative Closest Point (ICP)

The primary baseline method that the paper positions itself against is the **Iterative Closest Point (ICP)** algorithm [Besl & McKay, 1992]. Additionally, the paper also compares with the **TPS-RPM** (Thin Plate Spline-Robust Point Matching) method of Chui and Rangarajan [2003] and the **Coherent Point Drift (CPD)** method of Myronenko et al. [2006], both of which are EM-based approaches. However, ICP serves as the foundational reference point for motivating the paper's contribution.

**[2 marks]**

---

## Limitation of the Baseline

ICP relies on **hard, one-to-one nearest-neighbour correspondences**: at each iteration, every model point is paired with its closest scene point, and a transformation is computed to minimise the sum of squared distances across all pairs. This creates three fundamental weaknesses: (1) **Sensitivity to outliers** — a single outlier can dramatically distort the nearest-neighbour pairings and corrupt the estimated transformation. (2) **Dependence on good initialisation** — because the nearest-neighbour assignment is greedy and local, ICP easily gets trapped in local minima when the initial alignment is poor. (3) **No handling of partial overlap** — when the model is not a proper subset of the scene (e.g. due to occlusion), the forced one-to-one correspondences introduce systematic errors.

**[2 marks]**

---

## How the Proposed Method Overcomes This Limitation

The GMM-based method avoids explicit correspondence estimation entirely. By representing both point sets as continuous Gaussian mixture densities and minimising the L2 distance between them, the method implicitly allows **soft, many-to-many correspondences** — every model point contributes to the density everywhere, weighted by how close it is to scene points through the Gaussian kernel. The L2 distance's bounded influence function provides **inherent robustness** to outliers without requiring an explicit outlier rejection mechanism. The multi-scale annealing strategy (decreasing $\sigma$) further helps escape local minima by starting with a smooth, broadly peaked cost landscape.

**[1 mark]**

---

## Condition Under Which the Paper's Method Would Not Outperform ICP

The GMM-based method would **not outperform ICP** when the following conditions are all met: (1) the two point sets are **already coarsely aligned** (good initialisation available), (2) there are **no outliers** and no missing data (complete overlap), and (3) the point sets are **relatively small**. In this clean scenario, ICP's hard correspondences are actually correct — nearest neighbours genuinely correspond — and ICP converges much faster because it uses closed-form solutions for rigid transformations at each step, whereas the GMM method must evaluate expensive pairwise Gauss transforms ($O(k \cdot n)$ per function evaluation) and run iterative L-BFGS-B optimisation at multiple scales. Additionally, if the two point sets have **very different sampling densities** (see Assumption 2 in Task 1.2), the density-based approach of the GMM method can degrade because density peaks no longer reflect shape geometry, while ICP's point-to-point matching is less affected by this. Furthermore, for **rigid-only** registration problems, the overhead of setting up TPS bases and optimising over a large parameter space is entirely unnecessary — ICP's direct SVD-based solution is analytically optimal for each correspondence set.
